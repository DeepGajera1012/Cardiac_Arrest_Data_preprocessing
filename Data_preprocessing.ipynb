```python
df2 = pd.read_csv("Cardiac_Arrest.csv")
```

```python
cols = list(df2.columns)
a, b = cols.index('subject_id'), cols.index('id')
cols[a], cols[b] = cols[b], cols[a]
df2 = df2[cols]
```

```python
df2['gender'].replace({'F':0, 'M':1}, inplace=True)
```

```python
df2.sort_values(by=['subject_id', 'admission_time'])
```

```python
df2['admission_time'] = pd.to_datetime(df2['admission_time'])
df2['date_of_birth'] = pd.to_datetime(df2['date_of_birth'])
```

```python
for i in tqdm(range(0,len(df2))):
    if (df2["label"][i]=="admission_weight(lbs.)"):
        a=0.453592*(df2['valuenum'][i])
        df2['valuenum'][i]=a
        df2['label'][i]="admission_weight(Kg)"
```

```python
df["admission_weight(Kg)"] =df["admission_weight(Kg)"].astype(float)
df["heart_rate(bpm)"] =df["heart_rate(bpm)"].astype(float)
df["arterial_blood_pressure_mean(mmHg)"] =df["arterial_blood_pressure_mean(mmHg)"].astype(float)
df["non_invasive_blood_pressure_systolic-NBPS(mmHg)"] =df["non_invasive_blood_pressure_systolic-NBPS(mmHg)"].astype(float)
df["non_invasive_blood_pressure_diastolic-NBPD(mmHg)"] =df["non_invasive_blood_pressure_diastolic-NBPD(mmHg)"].astype(float)
```

```python
for i in tqdm(range(0,len(df))):
    l=[]
    l1=[]
    l2=[]
    l3=[]
    l4=[]
    for j in range(0,len(df2)):
        if (df2["id"][j] == df["id"][i]):
            if(df2.loc[j]['label'] == 'admission_weight(Kg)'):
                l.append(df2.loc[j]['valuenum'])
                m= np.mean(l)
                df["admission_weight(Kg)"][i] = m
            elif(df2.loc[j]['label'] == 'Heart Rate'):
                l1.append(df2.loc[j]['valuenum'])
                m1= np.mean(l1)
                df["heart_rate(bpm)"][i] = m1
            elif(df2["label"][j].__contains__("Arterial Blood Pressure")):
                l2.append(df2.loc[j]['valuenum'])
                m2= np.mean(l2)
                df["arterial_blood_pressure_mean(mmHg)"][i] = m2
            elif(df2["label"][j].__contains__("Non Invasive Blood Pressure systolic")):
                l3.append(df2.loc[j]['valuenum'])
                m3= np.mean(l3)
                df["non_invasive_blood_pressure_systolic-NBPS(mmHg)"][i] = m3
            elif(df2["label"][j].__contains__("Non Invasive Blood Pressure diastolic")):
                l4.append(df2.loc[j]['valuenum'])
                m4= np.mean(l4)
                df["non_invasive_blood_pressure_diastolic-NBPD(mmHg)"][i] = m4
```

```python
import math
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn import svm
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import multilabel_confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df['diagnosis'] = lb.fit_transform(df['diagnosis'])
df['language'] = lb.fit_transform(df['language'])
df['marital_status'] = lb.fit_transform(df['marital_status'])
df['religion'] = lb.fit_transform(df['religion'])
df['ethnicity'] = lb.fit_transform(df['ethnicity'])
```

```python
sns.countplot(df['target'])
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## LogisticRegression

```python
log =LogisticRegression(random_state = 0)
```

```python
log.fit(x_train,y_train)
```

```python
y_pred=log.predict(x_test)
```

```python
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
  
print ("Confusion Matrix : \n", cm)
```

## Random Forest Classifier

```python
RFC = RandomForestClassifier(n_estimators=300, max_depth=2, random_state=0)

RFC.fit(x_train, y_train)
```
